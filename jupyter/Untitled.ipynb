```python
import pandas as pd
```

```python
data = pd.read_csv('../data/CNAE-9.csv', names = ['category']+ [ "w_%s" % i for i in range(0, 856)])
```

```python
y = data.category
X = data[[ "w_%s" % i for i in range(0, 856)]]
print(y)
print(X)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X,y)
```

```python
pred = clf.predict(X)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y, pred)
```

```python
t = X.transpose()
t.to_csv('../data/CNAE-9-wide.csv')
```
